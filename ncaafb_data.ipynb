Creating tables for Venues, conferences, divisions

```python
tables_sql = [
    """CREATE TABLE IF NOT EXISTS venues (
        venue_id CHAR(36) PRIMARY KEY,
        name VARCHAR(150),
        city VARCHAR(100),
        state VARCHAR(100),
        country VARCHAR(100),
        zip VARCHAR(20),
        address VARCHAR(255),
        capacity INT,
        surface VARCHAR(50),
        roof_type VARCHAR(50),
        latitude DECIMAL(9,6),
        longitude DECIMAL(9,6)
    );""",

    """CREATE TABLE IF NOT EXISTS conferences (
        conference_id CHAR(36) PRIMARY KEY,
        name VARCHAR(100),
        alias VARCHAR(50)
    );""",

    """CREATE TABLE IF NOT EXISTS divisions (
        division_id CHAR(36) PRIMARY KEY,
        name VARCHAR(100),
        alias VARCHAR(50)
    );"""
]

for sql in tables_sql:
    cursor_pymysql.execute(sql)

conn_pymysql.commit()
print("Venues, Conferences, Divisions tables created")
```

Creating tables for Teams

```python
create_teams_sql = """
CREATE TABLE IF NOT EXISTS teams (
    team_id CHAR(36) PRIMARY KEY,
    market VARCHAR(100),
    name VARCHAR(100),
    alias VARCHAR(50),
    founded INT,
    mascot VARCHAR(100),
    fight_song VARCHAR(255),
    championships_won INT,
    conference_id CHAR(36),
    division_id CHAR(36),
    venue_id CHAR(36)
);
"""

cursor_pymysql.execute(create_teams_sql)
conn_pymysql.commit()
print("teams table created")
```

```python
Creating tables for Seasons, Players , Stats
```

```python
tables_sql = [
    """CREATE TABLE IF NOT EXISTS seasons (
        season_id CHAR(36) PRIMARY KEY,
        year INT NOT NULL,
        start_date DATE,
        end_date DATE,
        status VARCHAR(50),
        type_code VARCHAR(50)
    );""",

    """CREATE TABLE IF NOT EXISTS players (
        player_id CHAR(36) PRIMARY KEY,
        first_name VARCHAR(100),
        last_name VARCHAR(100),
        abbr_name VARCHAR(100),
        birth_place VARCHAR(150),
        position VARCHAR(50),
        height INT,
        weight INT,
        status VARCHAR(50),
        eligibility VARCHAR(50),
        team_id CHAR(36),
        FOREIGN KEY (team_id) REFERENCES teams(team_id)
    );""",

    """CREATE TABLE IF NOT EXISTS Stats (
        stat_id INT PRIMARY KEY,
        player_id CHAR(36),
        team_id CHAR(36),
        season_id CHAR(36),
        games_played INT,
        games_started INT,
        rushing_yards INT,
        rushing_touchdowns INT,
        receiving_yards INT,
        receiving_touchdowns INT,
        kick_return_yards INT,
        fumbles INT,
        FOREIGN KEY (player_id) REFERENCES players(player_id),
        FOREIGN KEY (team_id) REFERENCES teams(team_id),
        FOREIGN KEY (season_id) REFERENCES seasons(season_id),
        UNIQUE KEY uq_player_team_season (player_id, team_id, season_id)
    );"""
]

for sql in tables_sql:
    cursor_pymysql.execute(sql)

conn_pymysql.commit()
print("Seasons, Players, Stats tables created")
```
